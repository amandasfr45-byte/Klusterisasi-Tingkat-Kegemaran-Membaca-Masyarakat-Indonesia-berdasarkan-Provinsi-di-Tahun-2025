# Klusterisasi Tingkat Kegemaran Membaca Masyarakat Indonesia berdasarkan Provinsi di Tahun 2025

## Latar Belakang

Membaca merupakan salah satu indikator penting dalam meningkatkan kualitas sumber daya manusia dan pembangunan pendidikan di Indonesia. Tingkat kegemaran membaca masyarakat dapat mencerminkan tingkat literasi, akses informasi, serta budaya belajar di suatu daerah. Semakin tinggi tingkat kegemaran membaca masyarakat, maka semakin besar peluang masyarakat untuk meningkatkan pengetahuan, keterampilan, dan kualitas hidupnya.

Berdasarkan data Badan Pusat Statistik (BPS), tingkat kegemaran membaca masyarakat Indonesia tahun 2025 memiliki perbedaan pada setiap provinsi. Beberapa provinsi memiliki tingkat kegemaran membaca yang tinggi, sedangkan provinsi lainnya masih berada pada kategori rendah. Perbedaan tersebut dipengaruhi oleh berbagai faktor, seperti akses pendidikan, fasilitas perpustakaan, perkembangan teknologi informasi, kondisi sosial ekonomi, dan budaya masyarakat.

Perbedaan karakteristik tingkat kegemaran membaca antarprovinsi menyebabkan perlunya dilakukan pengelompokan wilayah agar dapat diketahui provinsi mana yang memiliki karakteristik serupa. Salah satu metode yang dapat digunakan untuk melakukan pengelompokan data adalah Data Mining dengan algoritma K-Means Clustering. Metode ini mampu mengelompokkan data berdasarkan kemiripan karakteristik sehingga hasilnya dapat digunakan untuk membantu pemerintah dalam menentukan kebijakan peningkatan literasi membaca masyarakat.

Melalui proses klusterisasi, pemerintah maupun instansi pendidikan dapat mengetahui kelompok provinsi dengan tingkat kegemaran membaca tinggi, sedang, maupun rendah. Hasil pengelompokan tersebut diharapkan dapat menjadi bahan evaluasi dan dasar pengambilan kebijakan dalam meningkatkan budaya literasi masyarakat Indonesia secara lebih efektif dan merata.

## Klasterisasi Tingkat Membaca/Literasi

Bagian ini bertujuan untuk mengelompokkan provinsi-provinsi berdasarkan tingkat kegemaran membaca mereka. Terdapat pendekatan yang digunakan dalam bagian ini:

### 1. Klasterisasi Manual Berdasarkan Rentang Nilai

Pada langkah ini, kolom baru 'Tingkat Literasi' dibuat dengan mengkategorikan 'Tingkat Kegemaran Membaca' ke dalam tiga level: 'Rendah', 'Sedang', dan 'Tinggi'. Pembagian ini didasarkan pada rentang nilai dari `min_value` hingga `max_value` yang dibagi menjadi tiga interval yang relatif sama. Pendekatan ini memberikan gambaran awal tentang distribusi tingkat literasi berdasarkan kriteria yang telah ditentukan secara manual.

```python
min_value = df_cleaned['Tingkat Kegemaran Membaca'].min()
max_value = df_cleaned['Tingkat Kegemaran Membaca'].max()

# Define bins for low, medium, and high reading levels
# Using roughly equal intervals based on the min and max values
bins = [min_value - 0.01, min_value + (max_value - min_value)/3, min_value + 2*(max_value - min_value)/3, max_value + 0.01]
labels = ['Rendah', 'Sedang', 'Tinggi']

# Create the new categorical column
df_cleaned['Tingkat Literasi'] = pd.cut(df_cleaned['Tingkat Kegemaran Membaca'], bins=bins, labels=labels, include_lowest=True)

# Display the counts for each category
display(df_cleaned['Tingkat Literasi'].value_counts())

display(df_cleaned.head())
```

## Klasterisasi Menggunakan K-Means Clustering

Metode *unsupervised machine learning* K-Means Clustering diterapkan untuk mengelompokkan provinsi secara objektif berdasarkan kemiripan karakteristik dari empat variabel utama:
*   'Tingkat Kegemaran Membaca'
*   'TGM-Pra Membaca'
*   'TGM-Saat Membaca'
*   'TGM-Pasca Membaca'

Langkah-langkah yang dilakukan adalah:
1.  **Penskalaan Fitur:** Data fitur distandarisasi menggunakan `StandardScaler` untuk memastikan semua variabel memiliki bobot yang setara dalam perhitungan jarak Euclidean. Ini penting karena K-Means sensitif terhadap skala data.
2.  **Penerapan K-Means:** Algoritma K-Means dijalankan dengan 3 klaster (`n_clusters=3`), mewakili 'Rendah', 'Sedang', dan 'Tinggi'. Parameter `random_state=42` digunakan untuk reproduktibilitas hasil, dan `n_init=10` untuk mendapatkan inisialisasi centroid terbaik.
3.  **Penetapan Label:** Setelah klasterisasi, label numerik (0, 1, 2) dari K-Means kemudian dipetakan ke label deskriptif 'Rendah', 'Sedang', dan 'Tinggi'. Pemetaan ini didasarkan pada nilai rata-rata 'Tingkat Kegemaran Membaca' di pusat masing-masing klaster, memastikan penamaan klaster yang logis.

Hasil dari K-Means Clustering ini diharapkan memberikan pengelompokan yang lebih robust dan berbasis data dibandingkan dengan klasterisasi manual, yang kemudian akan digunakan untuk analisis lebih lanjut dan rekomendasi kebijakan.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

# Features for clustering
features = ['Tingkat Kegemaran Membaca', 'TGM-Pra Membaca', 'TGM-Saat Membaca', 'TGM-Pasca Membaca']
X = df_cleaned[features]

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10) # n_init is set to suppress warning
df_cleaned['Cluster_KMeans'] = kmeans.fit_predict(X_scaled)

# Display the counts for each cluster
display(df_cleaned['Cluster_KMeans'].value_counts().sort_index())

# To assign meaningful labels (Rendah, Sedang, Tinggi), we can inspect the cluster centers
# For simplicity, we'll assume higher 'Tingkat Kegemaran Membaca' corresponds to higher literacy
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_df = pd.DataFrame(cluster_centers, columns=features)
cluster_df['Cluster'] = [0, 1, 2]
cluster_df = cluster_df.sort_values(by='Tingkat Kegemaran Membaca').reset_index(drop=True)

# Map the sorted clusters to 'Rendah', 'Sedang', 'Tinggi'
cluster_mapping = {
    cluster_df.loc[0, 'Cluster']: 'Rendah',
    cluster_df.loc[1, 'Cluster']: 'Sedang',
    cluster_df.loc[2, 'Cluster']: 'Tinggi'
}

df_cleaned['Tingkat Literasi_KMeans'] = df_cleaned['Cluster_KMeans'].map(cluster_mapping)

# Display the counts for the new labeled clusters
display(df_cleaned['Tingkat Literasi_KMeans'].value_counts())

# Display the head of the DataFrame with the new cluster assignments
display(df_cleaned.head())
```

```python
import geopandas
import folium
import matplotlib.colors as mcolors
import requests
import json

# URL to a GeoJSON file of Indonesian provinces (this might need to be updated if the source changes)
geoj_url = "https://raw.githubusercontent.com/superpikar/indonesia-geojson/master/indonesia.geojson"

try:
    response = requests.get(geoj_url)
    response.raise_for_status() # Raise an exception for HTTP errors
    indonesia_geojson = json.loads(response.text)
    print("GeoJSON data downloaded successfully.")
except requests.exceptions.RequestException as e:
    print(f"Error downloading GeoJSON: {e}")
    indonesia_geojson = None


if indonesia_geojson:
    # Load the GeoJSON data into a GeoDataFrame
    gdf = geopandas.GeoDataFrame.from_features(indonesia_geojson['features'])

    # Display the first few rows and check column names
    print("\nGeoDataFrame head:")
    display(gdf.head())
    print("\nUnique province names in GeoDataFrame (using 'state' column):")
    print(gdf['state'].unique())

    # Clean and standardize province names in the GeoDataFrame if necessary
    # Example of a common discrepancy: 'Daerah Istimewa Yogyakarta' vs 'DI Yogyakarta'
    # This mapping will depend on the exact names in the GeoJSON and your df_cleaned
    province_name_mapping = {
        'Jakarta Raya': 'DKI Jakarta',
        'Yogyakarta': 'DI Yogyakarta', # Assuming 'DI Yogyakarta' in your df_cleaned
        'Bangka-Belitung': 'Kepulauan Bangka Belitung', # GeoJSON name vs df_cleaned name
        # Add other mappings if discrepancies are found after inspecting unique names
    }
    gdf['NAME_1_cleaned'] = gdf['state'].replace(province_name_mapping)

    # Display unique cleaned names to verify
    print("\nUnique cleaned province names in GeoDataFrame:")
    print(gdf['NAME_1_cleaned'].unique())

    # Merge df_cleaned with gdf
    # Before merging, ensure 'Provinsi' in df_cleaned matches 'NAME_1_cleaned' in gdf
    # Remove the 'Indonesia' row from df_cleaned as it's an aggregate and not a province
    df_cleaned_provinces = df_cleaned[df_cleaned['Provinsi'] != 'Indonesia'].copy()

    # Check for unmatched provinces before merging
    gdf_provinces = set(gdf['NAME_1_cleaned'].unique())
    df_provinces = set(df_cleaned_provinces['Provinsi'].unique())

    print(f"\nProvinces in df_cleaned not in GeoJSON: {df_provinces - gdf_provinces}")
    print(f"Provinces in GeoJSON not in df_cleaned: {gdf_provinces - df_provinces}")

    # Merge the dataframes
    merged_gdf = gdf.set_index('NAME_1_cleaned').join(df_cleaned_provinces.set_index('Provinsi'))

    # Display merged_gdf info to check for successful merge and NaN values from unmatched provinces
    print("\nMerged GeoDataFrame info:")
    merged_gdf.info()
    display(merged_gdf.head())
else:
    print("Cannot proceed with map visualization without GeoJSON data.")
```

## Kesimpulan dan Interpretasi Hasil Klasterisasi

Setelah melakukan analisis dan klasterisasi menggunakan K-Means, kita dapat menarik beberapa kesimpulan mengenai tingkat kegemaran membaca di berbagai provinsi di Indonesia.

```python
# Tampilkan rata-rata 'Tingkat Kegemaran Membaca' per klaster

# Ensure df_cleaned is defined. This part re-runs necessary data loading and cleaning steps.
# It's recommended to run all cells above for a complete and clean execution flow.
import pandas as pd

# Re-define nama_file and df if they might not be in scope
nama_file = '/content/Tingkat Kegemaran Membaca Masyarakat dan Variabel Penyusunnya Menurut Provinsi, 2025.csv'
df = pd.read_csv(nama_file)

columns_to_clean = ['Tingkat Kegemaran Membaca', 'TGM-Pra Membaca', 'TGM-Saat Membaca', 'TGM-Pasca Membaca']

for col in columns_to_clean:
    df[col] = df[col].replace('–', pd.NA) # Replace '–' with pandas Not-Available marker

# Convert columns to numeric, coercing errors will turn non-convertible values into NaN
for col in columns_to_clean:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Drop rows where any of the specified columns have NaN values
df_cleaned = df.dropna(subset=columns_to_clean).reset_index(drop=True)

# Re-apply K-Means clustering and labels to df_cleaned to ensure 'Tingkat Literasi_KMeans' is present
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

features = ['Tingkat Kegemaran Membaca', 'TGM-Pra Membaca', 'TGM-Saat Membaca', 'TGM-Pasca Membaca']
X = df_cleaned[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_cleaned['Cluster_KMeans'] = kmeans.fit_predict(X_scaled)

cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_df = pd.DataFrame(cluster_centers, columns=features)
cluster_df['Cluster'] = [0, 1, 2]
cluster_df = cluster_df.sort_values(by='Tingkat Kegemaran Membaca').reset_index(drop=True)

cluster_mapping = {
    cluster_df.loc[0, 'Cluster']: 'Rendah',
    cluster_df.loc[1, 'Cluster']: 'Sedang',
    cluster_df.loc[2, 'Cluster']: 'Tinggi'
}

df_cleaned['Tingkat Literasi_KMeans'] = df_cleaned['Cluster_KMeans'].map(cluster_mapping)

display(df_cleaned.groupby('Tingkat Literasi_KMeans')['Tingkat Kegemaran Membaca'].mean().sort_values())
```

### Visualisasi Profil Klaster (Radar Chart)

Untuk memahami lebih jauh karakteristik masing-masing klaster, kita akan memvisualisasikan rata-rata nilai dari variabel pembentuk klaster ('Tingkat Kegemaran Membaca', 'TGM-Pra Membaca', 'TGM-Saat Membaca', 'TGM-Pasca Membaca') untuk setiap klaster dalam bentuk radar chart.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Pastikan cluster_df sudah diurutkan berdasarkan Tingkat Kegemaran Membaca
# (Seperti yang sudah dilakukan pada sel klasterisasi sebelumnya)
# cluster_df = cluster_df.sort_values(by='Tingkat Kegemaran Membaca').reset_index(drop=True)

# Rename the original cluster numbers to their meaningful labels for plotting
cluster_df['Tingkat Literasi'] = cluster_df['Cluster'].map(cluster_mapping)

# Prepare data for radar chart
# We'll use the 'cluster_df' which contains the inverse-transformed centroids
data_for_radar = cluster_df.set_index('Tingkat Literasi')[features]

# Number of variables
num_vars = len(features)

# Calculate angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot is a circle, so we need to "close" the plot by adding the first angle to the end
angles.append(angles[0])

# Create the radar chart
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot each cluster
for index, row in data_for_radar.iterrows():
    if index == 'Rendah':
        color = 'red'
    elif index == 'Sedang':
        color = 'orange'
    elif index == 'Tinggi':
        color = 'green'
    else:
        color = 'gray' # Fallback for unexpected index

    # The data for plotting must also be 'closed' by appending the first value to the end
    plot_data = np.concatenate((row.values, [row.values[0]]))

    ax.plot(angles, plot_data, color=color, linewidth=2, label=index)
    ax.fill(angles, plot_data, color=color, alpha=0.25)

# Set labels for each axis
ax.set_xticks(angles[:-1])
ax.set_xticklabels(features, fontsize=10)

# Set y-axis limits and labels
ax.set_ylim(40, 70) # Adjust limits based on your data range for better visualization
ytick_labels = np.arange(40, 70, 5)
ax.set_yticks(ytick_labels)
ax.set_yticklabels([str(label) for label in ytick_labels], color="gray", size=8)

# Add title and legend
ax.set_title('Profil Klaster Tingkat Kegemaran Membaca', size=16, color='black', y=1.08)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
ax.grid(True)

plt.show()
```

*   **Program Literasi Inklusif:** Luncurkan program literasi yang menargetkan kelompok usia dini dan masyarakat marginal dengan pendekatan yang menarik dan mudah diakses.
*   **Pengadaan Akses Buku:** Perbanyak perpustakaan umum, taman bacaan, dan sudut baca di fasilitas publik, serta distribusikan buku-buku yang relevan dan menarik secara gratis atau subsidi.
*   **Kampanye Kesadaran:** Lakukan kampanye besar-besaran untuk meningkatkan kesadaran akan pentingnya membaca melalui media lokal, acara komunitas, dan keterlibatan tokoh masyarakat.
*   **Pelatihan Guru/Pustakawan:** Berikan pelatihan kepada guru dan pustakawan tentang metode pengajaran membaca yang inovatif dan cara mengelola sumber daya perpustakaan secara efektif.

### Klaster 'Sedang'

Provinsi-provinsi ini memiliki tingkat kegemaran membaca yang moderat. Strategi di sini bertujuan untuk mempertahankan momentum dan mendorong peningkatan lebih lanjut:

*   **Pengembangan Kurikulum:** Integrasikan kegiatan membaca interaktif dan proyek berbasis literasi ke dalam kurikulum sekolah.
*   **Kemitraan Strategis:** Bangun kemitraan antara pemerintah daerah, sekolah, komunitas, dan sektor swasta untuk menyelenggarakan kegiatan literasi bersama, seperti festival buku, klub membaca, dan lokakarya penulisan.
*   **Digitalisasi Perpustakaan:** Kembangkan dan promosikan akses ke platform bacaan digital dan e-book untuk menyesuaikan dengan tren teknologi dan preferensi generasi muda.
*   **Pemberian Penghargaan:** Berikan insentif atau penghargaan kepada individu, sekolah, atau komunitas yang menunjukkan peningkatan signifikan dalam aktivitas membaca.

### Klaster 'Tinggi'

Klaster ini terdiri dari provinsi-provinsi dengan tingkat kegemaran membaca yang sudah tinggi. Fokusnya adalah pada keberlanjutan, inovasi, dan berbagi praktik terbaik:

*   **Inovasi Literasi:** Dorong inovasi dalam program literasi, seperti penggunaan teknologi augmented reality/virtual reality (AR/VR) untuk pengalaman membaca yang imersif, atau pengembangan platform penulisan kreatif.
*   **Pusat Keunggulan Literasi:** Jadikan provinsi-provinsi ini sebagai pusat keunggulan (center of excellence) literasi yang dapat menjadi percontohan dan tempat studi banding bagi provinsi lain.
*   **Penelitian dan Pengembangan:** Dukung penelitian untuk memahami faktor-faktor kunci keberhasilan dan tantangan yang masih ada, serta untuk mengembangkan strategi literasi masa depan.
*   **Promosi Budaya Membaca Global:** Ikut serta dalam inisiatif literasi internasional dan promosikan budaya membaca Indonesia di kancah global.

2. Bagaimana penerapan metode K-Means Clustering dalam mengelompokkan provinsi di Indonesia berdasarkan tingkat kegemaran membaca masyarakat?

Penyelesaian:

Penerapan metode K-Means Clustering dalam mengelompokkan provinsi di Indonesia berdasarkan tingkat kegemaran membaca masyarakat melibatkan langkah-langkah berikut:

1.  **Pemilihan Fitur (Features Selection):** Kami memilih empat variabel utama yang relevan dengan tingkat kegemaran membaca sebagai fitur untuk klasterisasi:
    *   `Tingkat Kegemaran Membaca`
    *   `TGM-Pra Membaca` (Tingkat Kegemaran Membaca sebelum membaca)
    *   `TGM-Saat Membaca` (Tingkat Kegemaran Membaca saat membaca)
    *   `TGM-Pasca Membaca` (Tingkat Kegemaran Membaca setelah membaca)

2.  **Penskalaan Fitur (Feature Scaling):** Sebelum menerapkan K-Means, fitur-fitur tersebut distandarisasi menggunakan `StandardScaler`. Penskalaan ini penting karena K-Means sangat sensitif terhadap skala data. Variabel dengan rentang nilai yang lebih besar akan memiliki pengaruh yang lebih dominan dalam perhitungan jarak antar titik, yang dapat bias hasil klasterisasi. `StandardScaler` mengubah data sehingga memiliki rata-rata nol dan variansi satu.

3.  **Penerapan K-Means:** Algoritma K-Means kemudian diterapkan pada data yang telah diskalakan (`X_scaled`). Jumlah klaster (`n_clusters`) ditetapkan menjadi 3, yang secara intuitif merepresentasikan tingkat 'Rendah', 'Sedang', dan 'Tinggi' dalam kegemaran membaca. `random_state=42` digunakan untuk memastikan hasil yang dapat direproduksi, dan `n_init=10` untuk menjalankan algoritma K-Means sebanyak 10 kali dengan inisialisasi centroid yang berbeda dan memilih hasil terbaik.

4.  **Penetapan Klaster ke Provinsi:** Setelah K-Means berjalan, setiap provinsi diberikan label klaster (0, 1, atau 2) berdasarkan centroid klaster terdekat.

5.  **Interpretasi Klaster:** Untuk memberikan makna pada label klaster numerik, kami memeriksa pusat klaster (`kmeans.cluster_centers_`) yang telah di-*inverse-transform* kembali ke skala aslinya. Dengan membandingkan nilai rata-rata `Tingkat Kegemaran Membaca` di setiap pusat klaster, kami dapat secara logis memetakan klaster dengan nilai terendah sebagai 'Rendah', klaster dengan nilai menengah sebagai 'Sedang', dan klaster dengan nilai tertinggi sebagai 'Tinggi'. Pemetaan ini dilakukan menggunakan `cluster_mapping`.

6.  **Analisis Hasil Klaster:** Hasil klasterisasi kemudian dianalisis melalui:
    *   **Statistik Deskriptif:** Menampilkan rata-rata dan statistik lainnya untuk setiap fitur di setiap klaster, memberikan gambaran profil masing-masing kelompok provinsi.
    *   **Daftar Provinsi:** Mendaftar provinsi-provinsi yang termasuk dalam setiap klaster.
    *   **Visualisasi:** Membuat berbagai visualisasi seperti:
        *   **Count plot:** Menunjukkan jumlah provinsi di setiap klaster.
        *   **Choropleth Map:** Memvisualisasikan klaster secara geografis di peta Indonesia untuk melihat pola spasial.
        *   **Radar Chart:** Menggambarkan profil rata-rata setiap klaster pada dimensi fitur yang digunakan.
        *   **Box Plot:** Menunjukkan distribusi setiap fitur dalam setiap klaster.
        *   **Scatter Plots dan Pair Plot:** Memvisualisasikan hubungan antar fitur dalam konteks klaster.

Melalui langkah-langkah ini, K-Means Clustering berhasil mengelompokkan provinsi-provinsi di Indonesia ke dalam kategori tingkat kegemaran membaca yang bermakna, memungkinkan identifikasi karakteristik dan perbedaan antara kelompok provinsi tersebut.
